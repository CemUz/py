```python
def generate_data(n):
    import numpy as np
    number_datapoints=n
    list_datapoints=[]

    for i in range(number_datapoints):
        datapoint=np.random.randn()
        list_datapoints.append(datapoint)
    return list_datapoints

data=generate_data(200)
plt.plot(data)
```

```python
def generate_data(n, data_distribution):
    import numpy as np
    number_datapoints=n
    list_datapoints=[]
    
    for i in range (number_datapoints):
        if data_distribution=='U':
            datapoint=np.random.uniform(0,1)
        else:
            datapoint=np.random.randn()
        list_datapoints.append(datapoint)
    
    return list_datapoints

import matplotlib.pyplot as plt
data=generate_data(100,'')
plt.plot(data)
```

```python
def computepay(hours, rate):
    
    try:
        
        if(int(hours)>=40):
            overtime=int(hours)-40
       
        else:
            overtime=0
        
        print('gross pay:',int(rate)*(40+(overtime*1.5)))

    except:
        
        print('please only enter integers')

        
hours=input('hours?',)
rate=input('rate?',)

x=computepay(hours, rate)
```

```python
def computergrade(score):
    
    try:
        
        score=float(score)
        
        if(1.0>=score>=0.9):
            print('A!')
        elif(0.9>score>=0.8):
            print('B!')
        elif(0.8>score>=0.7):
            print('C!')
        elif(0.7>score>=0.6):
            print('D!')
        elif(0.0<=score<0.6):
            print('F!')
        else:
            print('pls only enter numbers between 0 and 1')
    except:
        print('bad grade!')
        
score=(input('score?',))
x=computergrade(score)
```
