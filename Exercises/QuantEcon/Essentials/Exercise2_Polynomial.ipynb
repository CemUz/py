```python
def polynomial(x, coeff):
        
    y=sum([coefficient*(x**index) for index, coefficient in enumerate(coeff)])

    return(y)
          
import numpy as np
import matplotlib. pyplot as plt

c=4
coeff=[]
    
for i in range(c+1):
    coeff.append(np.random.uniform(0.5,2))
    
x=0
datalist=[]

for i in range(100):
    data=polynomial(x, coeff)
    datalist.append(data)
    x+=1
    
plt.plot(datalist)
```
