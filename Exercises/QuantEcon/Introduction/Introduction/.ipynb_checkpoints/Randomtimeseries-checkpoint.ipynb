```python
'''Here, differential equations are plotted.
Every (absolute) value x-1 is weighed with alpha and a random value is added to generate x.
T describes the total number of values generated for each alpha.'''

import numpy as np
import matplotlib.pyplot as plt

T=50
alpha=[0.0, 0.9, 0.98]

x=0

for i in alpha:
    
    a=float(i)
    print(i, a)
    xvalues=[0]
    
    for i in range(T):
    
        e=np.random.randn()
        x=a*abs(x)+e
        if x<0:
            x=-x
        xvalues.append(x)
    
    plt.plot(xvalues, label='a = '+str(a))
    
plt.legend(frameon=False)
```
