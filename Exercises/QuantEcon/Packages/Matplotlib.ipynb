```python
from scipy.stats import norm
from random import uniform

fig, ax = plt.subplots()
x = np.linspace(-4, 4, 150)

for i in range(3):
    m, s = uniform(-1,1), uniform(1,2)
    y = norm.pdf(x, loc=m, scale=s)
    current_label = f'$\sigma = {s:.2},\mu = {m:.2}$'
    ax.plot(x,y,'.',lw=2, alpha=0.5, label=current_label)
    
ax.legend()
```

```python
num_rows, num_cols = 3, 2
fig, axes = plt.subplots(num_rows, num_cols, figsize=(10,12))

for i in range(num_rows):
    for j in range(num_cols):
        m, s = uniform(-1,1), uniform(1,2)
        x = norm.rvs(loc=m, scale=s, size=100)
        
        axes[i, j].hist(x, alpha=0.6, bins=20)
        
        t = f'$\mu = {m:.2}, \quad \sigma = {s:.2}$'
        axes[i, j].set(title=t, xticks=[-4,0,4], yticks=[])
        
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib  import cm

def f(x,y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

xgrid = np.linspace(-4, 4, 500)
ygrid = xgrid
x, y = np.meshgrid(xgrid, ygrid)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x,
                y,
                f(x, y),
                rstride=2, cstride=2,
                cmap=cm.jet,
                alpha=0.7,
                lw=0.25)

ax.set_zlim(-0.5,1.0)
```

```python
import matplotlib.ticker as tck

def subplots():
    fig, ax = plt.subplots()
    
    for spine in ['left', 'bottom']:
        ax.spines[spine].set_position('zero')
    for spine in ['right', 'top']:
        ax.spines[spine].set_color('none')
        
    ax.xaxis.set_major_formatter(tck.FormatStrFormatter('%g $\pi$'))
    ax.xaxis.set_major_locator(tck.MultipleLocator(base=1.0))
    
    return fig, ax

fig, ax = subplots()

x = np.linspace(-np.pi,np.pi,100)
y = np.sin(x)

ax.plot(x/np.pi,y,'r.',alpha=0.6,label='$y = \sin(x)$')
ax.set(yticks=[-1,0,1], xticks=[-1, 0, 1], title='Zentriertes Koordinatensystem \nmit grenzen von $-\pi$ bis $\pi$')

ax.legend()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
from matplotlib import cm

fig, ax = plt.subplots()

t = np.linspace(0,5,20)
x = np.linspace(0,3,100)
T, X = np.meshgrid(t, x)

y = np.cos(np.pi*T*X)*np.exp(-X)

ax.plot(X, y, alpha=0.8)
plt.show()

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,
                T,
                y,
                rstride=2, cstride=2,
                cmap=cm.viridis,
                alpha=0.6,
                linewidth=0.25)
ax.set_zlim(-0.5, 0.8)
plt.show()
```
