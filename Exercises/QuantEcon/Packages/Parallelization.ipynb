```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from numba import njit, vectorize

@vectorize('float64(float64, float64)', target='parallel')
def f(x, y):
    return np.cos(x**2 + y**2) / (1 + x**2 + y**2)

grid = np.linspace(-3, 3, 5000)
x, y = np.meshgrid(grid, grid)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot_surface(x, y, f(x,y))

%timeit np.max(f(x,y))
```

```python
import numpy as np
from numpy.random import randn
from numba import njit

@njit
def h(w, r=0.1, s=0.3, v1=0.1, v2=1.0):
    """
    Updates household wealth.
    """

    # Draw shocks
    R = np.exp(v1 * randn()) * (1 + r)
    y = np.exp(v2 * randn())

    # Update wealth
    w = R * s * w + y
    return w

@njit
def compute_long_run_median(w0=1, T=1000, num_reps=50_000):
    
    obs = np.empty(num_reps)
    
    for i in range(num_reps):
        w = w0
        for j in range(T):
            w = h(w)
        obs[i] = w
        
    return np.median(obs)
```

```python
%%time
compute_long_run_median()
```

```python
from numpy.random import randn
from numba import njit, prange
import numpy as np

@njit
def hp(w, r=0.1, s=0.3, v1=0.1, v2=1.0):
    """
    Updates household wealth.
    """

    # Draw shocks
    R = np.exp(v1 * randn()) * (1 + r)
    y = np.exp(v2 * randn())

    # Update wealth
    w = R * s * w + y
    return w

@njit(parallel=True)
def compute_long_run_median_parallel(w0=1, T=1000, num_reps=50_000):
    
    obs = np.empty(num_reps)
    
    for i in prange(num_reps):
        w = w0
        for j in range(T):
            w = hp(w)
        obs[i] = w
        
    return np.median(obs)
```

```python
%%time
compute_long_run_median_parallel()
```

```python
import numpy as np
from numba import njit, prange
import matplotlib.pyplot as plt

'''generating a set of paired values representing randomly distributed points in a quadratic 2D plane
and calculating the length of the corresponding vectors. 
Additionally, calculating the number of points IN a (TECHNICALLY, quarter) of a cricle(vectorlength<R)originating at 0,0.
Later, we will be using the fraction of points in the circle to determine the probability of a point to fall into the circle and,
thus in turn, the area of the circle.'''

@njit(parallel=True)
def generate_pairs(numberofvectors):
    
    '''successively generate random vectors in a 2D-plane(0 to 1) and determine its distance from its origin(0,0).
    If points are more than one away from origin, then its not counted as being inside the circle.
    from the fraction of points inside, one can calculate the area of the circle relative to the total area (which is 1)'''
    
    inside=0
    out=0
    
    for j in prange(numberofvectors):
        
        x_coord=np.random.uniform(0,1)
        y_coord=np.random.uniform(0,1)

        vectorlength=np.sqrt((x_coord**2)+(y_coord**2))
        if(vectorlength<1):
            inside=inside+1   
        else:
            out=out+1
                 
    quarterarea=inside/(out+inside)
    area=4*quarterarea

    return(area)


'''define a function which enables to set a number of times the generate_pairs function is called
and at each call remember the result by appending the result to a list'''

@njit
def generate_trials(numberoftrials, numberofpairs):
    
    pi_list=[]
    
    for i in range (numberoftrials):
        
        pivalue=generate_pairs(numberofpairs)
        pi_list.append(pivalue)
        
    return(pi_list)

%time pi = generate_trials(1, 1000000000)
plt.plot(pi)
np.mean(pi)
```
