```python
def binomial_rv(p, n):
    
    '''binomial_rv defines a probability p as a fraction from a set of uniformly distributed numbers below a given limit (p!). 
    This is done by first assigning the variable u to a random value between 0 and 1. 
    If the value is inside the given limit between 0 and p, then a counter is increasing by one. 
    For a given number of draws n, this delivers an approximation to p, which is converging to p for a big amount of draws n.'''
    
    success=0
    fail=0
    
    for i in range(n):
        
        u=uniform(0,1)
        
        if(u<=p):
            success=success+1
            
    successfraction=success/n
    
    return(successfraction)   
```

```python
def many_trials(trialno):
    
    '''many_trials loops through binomial_rv a number of times and at each end of the loop appends the result to a list 'trials'.
    this is done a 'trialno' of times.'''
    
    trials=[]
    
    for i in range(trialno):
        
        data=binomial_rv(p, n)
        trials.append(data)
        
    return(trials)
```

```python
def trialnodependency(maxtrialno):
    
    '''trialnodependency assigns a 'trials' list to the variable 'zdata', plotted at y=1 and x=n.
    two counters make sure that the next datalist(hence the next 'trials' list) is plotted at y=2 and x=n+1 and so on.
    n is set to global due to the fact that n, a variable used outside the function(number of draws in binomial_rv),
    is increased by one at every end of a loop INSIDE the function.'''
    
    global n
    
    for i in range(maxtrialno):

        zdata = many_trials(trialno)
        xdata = n
        ydata = list(range(1,trialno+1))

        ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='cividis') 
        n=n+1
        
    return(fig)
```

```python
'''Variables (p, n, trialno(amount of lists), maxtrialno(result for the highest amount of draws plotted) can be set here,
also parameters for the plot can be set, like figure size, axis orientation etc. 
In the end, the algorithm is called by assigning fig to trialnodependency(maxtrialno)'''

%matplotlib notebook
%matplotlib notebook

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from numpy.random import uniform

p=0.5
n=1
trialno=10
maxtrialno=50

fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
ax.invert_xaxis()

fig=trialnodependency(maxtrialno)



```
