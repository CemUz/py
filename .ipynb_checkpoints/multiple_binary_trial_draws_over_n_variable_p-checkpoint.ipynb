```python
def binomial_rv(p, n):
    
    success=0
    fail=0
    
    for i in range(n):
        
        u=uniform(0,1)
        
        if(u<=p):
            success=success+1
            
        else:
            fail=fail+1
            
    successfraction=success/n
    
    return(successfraction)   
```

```python
def many_trials(trialno):
    
    trials=[]
    
    for i in range(trialno):
        
        data=binomial_rv(p, n)
        trials.append(data)
        
    return(trials)
```

```python
def trialnodependency(maxtrialno):
    
    global n
    
    for i in range(maxtrialno):
        
        dataset=many_trials(trialno)

        zdata = dataset
        xdata = n
        ydata = list(range(1,trialno+1))

        ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='cividis') 
        n=n+1
        
    return(fig)
```

```python
%matplotlib inline
from mpl_toolkits import mplot3d
from numpy.random import uniform


p=0.3
n=1
trialno=400
maxtrialno=200

fig = plt.figure(figsize=(24,18))
ax = plt.axes(projection='3d')
ax.invert_xaxis()

fig=trialnodependency(maxtrialno)


```
