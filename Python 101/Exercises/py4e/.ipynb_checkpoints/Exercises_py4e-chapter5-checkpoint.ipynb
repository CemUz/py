```python
'''Make a list of inputs repeatedly entered by the user until the user enters 'done'. 
Adding and counting the items as well as calculating the average is done afterwards. 
In the end, the script prints the values for sum, count and average.'''

def calc():
    '''Make a list of numbers entered by the user until he terminates the process by entering 'done'
    If the user enters invalid values (e.g. words), he will be notified to enter numbers'''
    
    list=[]
    
    while True:
        number=input('Enter a number:',)
        if (number=='done'):
            break
        else:
            try:
                list.append(int(number))
            except:
                print('Invalid value. Only enter digits')
                
    return(list)

'''Call calc() to attain a user-generated list of numbers.
Use the list generated by calc() to calculate sum, count and average of items from list'''

list=calc()

if(len(list)==0):
    print('No values!')
else:
    print('Sum:',sum(list))
    print('Count:',len(list))
    print('Average:',sum(list)/len(list))
    print('Largest:',max(list))
    print('smallest:',min(list))
```
