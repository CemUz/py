```python
#generating a set of paired values representing randomly distributed points in a quadratic 2D plane
#and calculating the length of the corresponding vectors. 
#Additionally, calculating the number of points IN a (TECHNICALLY, quarter) of a cricle(vectorlength<R) originating at 0,0.
#later, using the fraction of points in the circle to determine the probability of a point to fall into the circle,
#which in turn equals the area of the circle. 

def generate_pairs(numberofvectors):
    
    import numpy as np
    
    inside=0
    out=0
    
    inxlist=[]
    inylist=[]
    outxlist=[]
    outylist=[]
    
    #successively generate random vectors in a 2D-plane(0 to 1) and determine its distance from its origin(0,0)
    for j in range(numberofvectors):
        
        #choose values for x and y
        x_coord=np.random.uniform(0,1)
        y_coord=np.random.uniform(0,1)
        
        #compute vectorlength and assign either to a or b (outside or inside of circle)
        vectorlength=np.sqrt((x_coord**2)+(y_coord**2))
        if(vectorlength<1):
            inside=inside+1
            inxlist.append(x_coord)
            inylist.append(y_coord)
            
        else:
            out=out+1
            outxlist.append(x_coord)
            outylist.append(y_coord)
            
    #compute and return the fraction of b        
    ratio_inout=inside/out
    return(inxlist, inylist, outxlist, outylist)
    

#define a function which enables to set a number of times the generate_pairs function is called    
def generate_trials(numberoftrials, numberofpairs):
    
    pi_list=[]
    
    #call the function generate_pairs a numberoftrials times and at each end append the result in pi_list
    for i in range (numberoftrials):
        
        pivalue=generate_pairs(numberofpairs)
        pi_list.append(pivalue)
    
    return(pi_list)

#call the function generate trials by defining number of trials and number of pairs.
#Calculate Average and plot every trial result.
import numpy as np
import matplotlib.pyplot as plt

pi=generate_trials(1, 100)
#print(np.mean(pi))
plt.plot(inxlist, inylist,'g')
plt.plot(outxlist, outylist, 'b')


```
